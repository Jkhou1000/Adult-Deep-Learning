```python
import tensorflow as tf
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
#Feature Data. Drop the target
X = adult.drop(columns='income')
#Target is income
y = adult['income']


#train_test_split
X_train, X_test, ytrain, ytest = train_test_split(X, y, random_state=42)
```

```python
scl = StandardScaler()
ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')
#Selectors
cats = make_column_selector(dtype_include='object')
nums = make_column_selector(dtype_include='number')
#tuple
num_tup = (scl, nums)
cat_tup = (ohe, cats)
transformer = make_column_transformer(num_tup, cat_tup, remainder='passthrough')
```

```python
X_train = transformer.fit_transform(X_train)
X_test = transformer.transform(X_test)
input_shape = X_train.shape[1]
Lenc = LabelEncoder()
y_train = Lenc.fit_transform(ytrain)
y_test = Lenc.transform(ytest)
```

```python
print(np.isnan(X_train).sum().sum(), 'missing values in training data')
print(np.isnan(X_test).sum().sum(), 'missing values in training data')
print('\n')
print('All data in Xtrain are', X_train.dtype)
print('All data in Xtest are', X_test.dtype)
print('\n')
print('data shape', X_train.shape)
print('data shape', X_test.shape)
```

```python
model = Sequential()
```

```python
model.add(Dense(100, activation='relu', input_dim= X_train.shape[1]))
model.add(Dense(58, activation= 'relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=7, verbose=1)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report

def class_eval(y_true, y_pred, labels=None, normalize=None):
    print(classification_report(y_true, y_pred, target_names=labels))

    ConfusionMatrixDisplay.from_predictions(y_true, y_pred,
    display_labels=labels,
    normalize=normalize,
    cmap='Blues')

plt.show()

def plot_hist(history, metrics=None):
    plt.plot(history.history['loss'], label='training')
    plt.plot(history.history['val_loss'], label='testing')
    plt.title('Loss')
    plt.legend()
    plt.show()
    if metrics:
        for metric in metrics:
            plt.plot(history.history[metric], label=f'training{metric}' )
            plt.plot(history.history[f'val_{metric}'], label=f'testing{metric}')
            plt.legend()
            plt.title(metric)
            plt.show()
```

```python
plot_hist(history, ['acc'])
```

```python
labels = ['<=50k', '>50k']

train_pred = np.rint(model.predict(X_train))
test_pred = np.rint(model.predict(X_test))

print('Training Evaluation:\n')
class_eval(y_train, train_pred, labels=labels, normalize='true')

print('Testing Evaluation:\n')
class_eval(y_test, test_pred, labels=labels, normalize='true')


```

### This model has a high accuracy when it comes to getting true positives but needs more tuning to get its true negatives. 

```python
model2 = Sequential()

model2.add(Dense(124, activation='relu', input_dim=X_train.shape[1]))
model2.add(Dropout(0.2))
model2.add(Dense(73, activation='relu'))
model2.add(Dropout(0.1))
model2.add(Dense(1, activation='sigmoid'))

model2.compile(loss='bce', optimizer='adam', metrics=['acc'])

hist2 = model2.fit(X_train, y_train,
validation_data=(X_test, y_test),
epochs=15)
```

```python
plot_hist(hist2, ['acc'])
```

```python
train_pred2 = np.rint(model2.predict(X_train))
test_pred2 = np.rint(model2.predict(X_test))

print('Training Evaluation:\n')
class_eval(y_train, train_pred2, labels=labels, normalize='true')
print('Testing Evaluation:\n')
class_eval(y_test, test_pred2, labels=labels, normalize='true')
```

### The Accuracy and F1 score have fluctuated a bit, True negatives are slightly better on the training data, the next model should be less complex again

```python
#Denser model with a lot of epochs
from keras import regularizers
from keras.callbacks import EarlyStopping
model3 = Sequential()

model3.add(Dense(200, activation='relu', input_dim=X_train.shape[1], kernel_regularizer=regularizers.l2(0.01)))
model3.add(Dropout(0.3))
model3.add(Dense(120, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model3.add(Dropout(0.2))
model3.add(Dense(30, activation='relu', ))
model3.add(Dropout(0.1))
model3.add(Dense(1, activation='sigmoid'))

model3.compile(loss='bce', optimizer='adam', metrics=['acc'])

hist3 = model3.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, callbacks=EarlyStopping(patience=20))
```

```python
plot_hist(hist3, ['acc'])


train_pred3 = np.rint(model3.predict(X_train))
test_pred3 = np.rint(model3.predict(X_test))

print('Training Evaluation\n')
class_eval(y_train, train_pred3, labels=labels, normalize='true')
print('Testing Evaluation\n')
class_eval(y_test, test_pred3, labels=labels, normalize='true')
```

### This marks the highest f1 score of all 3 models, barely beating out model 2's testing f1 of .68 at an f1 of .7 on this model. Further iterations may require more regularization and/or less complex layers. The accuracy also unflinchingly remained around .86-.88 on all 3 models for this kernel.
